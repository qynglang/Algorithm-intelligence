```python
N='27'
ans=0
E1=np.load('Data/'+N+str('E1_distributed')+'.npy')
E1_1=np.load('Data/'+N+str('E1_1_distributed')+'.npy')
E1_2=np.load('Data/'+N+str('E1_2_distributed')+'.npy')
E1_3=np.load('Data/'+N+str('E1_3_distributed')+'.npy')
E1_4=np.load('Data/'+N+str('E1_4_distributed')+'.npy')
E_min_n=np.zeros((50))
for i in range (0,10):
    E_min_n[i]=np.argmin(E1[i])
for i in range (10,20):
    E_min_n[i]=np.argmin(E1_1[i-10])
for i in range (20,30):
    E_min_n[i]=np.argmin(E1_2[i-20])
for i in range (30,40):
    E_min_n[i]=np.argmin(E1_1[i-30])
for i in range (40,50):
    E_min_n[i]=np.argmin(E1_1[i-40])
print(E_min_n[E_min_n==ans].size)
print(np.average(E1,axis=0))
print(np.argsort(np.average(E1,axis=0)))
print(np.average(E1_1,axis=0))
print(np.argsort(np.average(E1_1,axis=0)))
print(np.average(E1_2,axis=0))
print(np.argsort(np.average(E1_2,axis=0)))
print(np.average(E1_3,axis=0))
print(np.argsort(np.average(E1_3,axis=0)))
print(np.average(E1_4,axis=0))
print(np.argsort(np.average(E1_4,axis=0)))
print(np.argsort(np.average(E1_4+E1_3+E1_2+E1_1+E1,axis=0)))
a=np.average(E1,axis=0)[ans]/E1.mean()
b=np.average(E1_1,axis=0)[ans]/E1_1.mean()
c=np.average(E1_2,axis=0)[ans]/E1_2.mean()
d=np.average(E1_3,axis=0)[ans]/E1_3.mean()
e=np.average(E1_4,axis=0)[ans]/E1_4.mean()
print((a+b+c+d+e)/5)
```
