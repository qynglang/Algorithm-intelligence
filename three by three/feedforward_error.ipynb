```python
from data_gene import forward
EE=np.zeros((8,5))
for j in range (0,5):
    II=forward(data[:,:,:,2],w1[:,:,:,:,j],w2[:,:,:,:,j])
    error01=II[1,:]-II[0,:]
    error12=II[2,:]-II[1,:]
    error34=II[4,:]-II[3,:]
    error54=II[5,:]-II[4,:]
    error67=II[7,:]-II[6,:]
    error78=np.zeros((8,294))
    for i in range (0,8):
        error78[i,:]=II[8+i,:]-II[7,:]
    O1=np.dot(error67.reshape(1,294),wf1[:,:,j])+bf1[:,j]
    O2=np.zeros((8,1024))
    for i in range (0,8):
        O2[i,:]=np.dot(error78[i,:].reshape(1,294),wf1[:,:,j])+bf1[:,j]
    E=np.zeros(8)
    for i in range (0,8):
        E[i]=abs(O2[i,:]-O1).sum()
    EE[:,j]=E
```

```python
ans=2
from data_gene import forward
EE=np.zeros((8,5))
for j in range (0,5):
    II=forward(data[:,:,:,2],w1[:,:,:,:,j],w2[:,:,:,:,j])
    error01=II[1,:]-II[0,:]
    error12=II[2,:]-II[1,:]
    error34=II[4,:]-II[3,:]
    error54=II[5,:]-II[4,:]
    error67=II[7,:]-II[6,:]
    error78=np.zeros((8,294))
    for i in range (0,8):
        error78[i,:]=II[8+i,:]-II[7,:]
    O1=np.dot(error67.reshape(1,294),wf1[:,:,j])+bf1[:,j]
    O2=np.zeros((8,1024))
    for i in range (0,8):
        O2[i,:]=np.dot(error78[i,:].reshape(1,294),wf1[:,:,j])+bf1[:,j]
    E=np.zeros(8)
    for i in range (0,8):
        E[i]=abs(O2[i,:]-O1).sum()
    EE[:,j]=E
    N=np.zeros(5)
    for i in range (0,5):
        N[i]=np.argmin(EE[:,i])
    print(np.argmin(np.average(EE,axis=1)))
    print(np.argsort(np.average(EE,axis=1)))
    print(np.average(EE,axis=1)[ans]/np.average(EE,axis=1).mean())
```
